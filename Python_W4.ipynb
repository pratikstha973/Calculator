Question 1 — Small Shop Billing and Inventory System

```python
inventory_shop = {
    "rice":  {"price": 120, "stock": 20},
    "milk":  {"price": 90,  "stock": 10},
    "bread": {"price": 60,  "stock": 15},
    "eggs":  {"price": 15,  "stock": 30},
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12,
}

def process_order(inventory, cart):
    bill_items = {}
    grand_total = 0

    for item, qty in cart.items():
        if item not in inventory:
            print(f"Sorry, {item} is not in inventory.")
            continue
        if inventory[item]["stock"] < qty:
            print(f"Sorry, not enough stock for {item}.")
            continue
        cost = inventory[item]["price"] * qty
        bill_items[item] = (qty, cost)
        grand_total += cost
        inventory[item]["stock"] -= qty

    print("\n---- Bill ----")
    for item, (qty, cost) in bill_items.items():
        print(f"{item} x{qty} = NPR {cost}")
    print(f"Grand Total: NPR {grand_total}")
    print("--------------")

    stock_summary = ", ".join(
        f"{item}={data['stock']}" for item, data in inventory.items()
    )
    print(f"Updated stock: {stock_summary}")

process_order(inventory_shop, cart)
```

Question 2 — Water Level Alert System (Koshi River)

```python
sensors = [
    ("Chatara",         2.8),
    ("Tribeni Ghat",    5.4),
    ("Koshi Barrage",   4.1),
    ("Sunsari Bridge",  1.9),
    ("Saptakoshi Camp", 6.0),
]

def check_water_level(location, level_metres):
    if level_metres < 3:
        return "Safe"
    elif level_metres <= 5:
        return "Warning — Alert nearby villages"
    else:
        return "DANGER — Evacuate immediately!"

for location, level in sensors:
    status = check_water_level(location, level)
    print(f"{location} ({level} m): {status}")
```

Question 3 — Date Converter for Nepal Bank System (BS ↔ AD)

```python
bs_months = ["Baisakh", "Jestha", "Ashadh", "Shrawan", "Bhadra", "Ashwin",
             "Kartik", "Mangsir", "Poush", "Magh", "Falgun", "Chaitra"]

def convert_date(date_str, from_cal, to_cal, style="iso"):
    year, month, day = map(int, date_str.split("-"))

    if from_cal == to_cal:
        converted_year, converted_month, converted_day = year, month, day
        converted_cal = to_cal
    elif from_cal == "AD" and to_cal == "BS":
        converted_year  = year + 56
        converted_month = month
        converted_day   = day
        converted_cal   = "BS"
    else:  # BS -> AD
        converted_year  = year - 56
        converted_month = month
        converted_day   = day
        converted_cal   = "AD"

    if style == "iso":
        return f"{converted_year:04d}-{converted_month:02d}-{converted_day:02d} {converted_cal}"
    elif style == "full":
        suffix = {1:"st", 2:"nd", 3:"rd"}.get(converted_day if converted_day < 20
                                                else converted_day % 10, "th")
        month_name = bs_months[converted_month - 1] if converted_cal == "BS" \
                     else converted_month
        return f"{converted_day}{suffix} {month_name}, {converted_year} {converted_cal}"
    elif style == "nepali":
        month_name = bs_months[converted_month - 1] if converted_cal == "BS" \
                     else converted_month
        return f"{converted_day} {month_name} {converted_year} {converted_cal}"
    else:
        return f"{converted_year:04d}-{converted_month:02d}-{converted_day:02d} {converted_cal}"

customers = [
    {"name": "Ramesh Thapa",  "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki",  "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai",    "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]

for c in customers:
    result = convert_date(c["date"], c["cal"], c["need"], c["style"])
    print(f"{c['name']} | Original: {c['date']} {c['cal']} | Converted: {result}")
```

Question 4 — Word Frequency Counter

```python
text = """
Nepal is a beautiful country. Nepal has Mount Everest.
Everest is the highest mountain in the world. Many tourists
visit Nepal every year to see Everest and other mountains.
Nepal is known for its mountains and natural beauty.
"""

def word_frequency(text):
    cleaned = ""
    for ch in text:
        if ch.isalpha() or ch.isspace():
            cleaned += ch.lower()
        else:
            cleaned += " "

    words = cleaned.split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    top3 = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:3]
    return top3

top_words = word_frequency(text)
print("Top 3 words:")
for word, count in top_words:
    print(f"  {word} — {count} times")
```

Question 5 — Simple ATM Simulator

```python
accounts = {
    "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki", "balance": 8500,  "pin": "5678"},
    "A003": {"name": "Bikash Rai",   "balance": 22000, "pin": "9012"},
}

def atm(account_id, pin, action, amount=0):
    if account_id not in accounts:
        print("Account not found.")
        return
    acc = accounts[account_id]
    if acc["pin"] != pin:
        print("Incorrect PIN.")
        return

    if action == "balance":
        print(f"{acc['name']} | Balance: NPR {acc['balance']}")
    elif action == "deposit":
        acc["balance"] += amount
        print(f"Deposited NPR {amount}. New balance: NPR {acc['balance']}")
    elif action == "withdraw":
        if amount > acc["balance"]:
            print("Insufficient funds.")
        else:
            acc["balance"] -= amount
            print(f"Withdrawn NPR {amount}. New balance: NPR {acc['balance']}")
    else:
        print(f"Unknown action: {action}")

atm("A001", "1234", "balance")
atm("A002", "0000", "withdraw", 2000)   # wrong PIN
atm("A002", "5678", "deposit",  3000)
atm("A003", "9012", "withdraw", 25000)  # insufficient funds
atm("A004", "1111", "balance")          # account not found
```
